```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('FilterAggregate').getOrCreate()

df_pyspark=spark.read.csv('test3.csv',header=True,inferSchema=True)
df_pyspark.show()
```

```python
# filter salary less than 20000
df_pyspark.filter("salary>=10000").show()

df_pyspark.filter("salary>=10000").select(['Name','salary']).show()

df_pyspark.filter(df_pyspark['salary']>=10000).show()

df_pyspark.filter((df_pyspark['salary']>=5000) | 
                  (df_pyspark['salary']<=20000)).show()

df_pyspark.filter(~(df_pyspark['salary']<=20000)).show()
df_pyspark.filter(~(df_pyspark['salary']<=15000)).show()
```

```python
# Aggregation

# find the maximum salary
df_pyspark.groupBy('Name').sum().show()

# finding average
df_pyspark.groupBy('Name').avg().show()

# lets find total salary by departments
df_pyspark.groupBy('Departments').sum().show()

# lets find average salary by departments
df_pyspark.groupBy('Departments').mean().show()

# lets find how many people working for a department
df_pyspark.groupBy('Departments').count().show()

# total salary paid by all departments
df_pyspark.agg({'salary':'sum'}).show()
```
